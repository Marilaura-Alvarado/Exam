```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Name of the student': ['Laura', 'Maria', 'Artem', 'Jimena', 'Sofia'],
    'Attendance (%)': [80, 99, 23, 60, 17],
    'Mark (points)': [90, 99, 45, 55, 35]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
plt.scatter(df['Attendance (%)'], df['Mark (points)'], color='blue', marker='o')
plt.title('Correlation between attendance and evaluation')
plt.xlabel('Attendance  (%)')
plt.ylabel('Mark (points)')
plt.grid(True)
plt.show()
```
